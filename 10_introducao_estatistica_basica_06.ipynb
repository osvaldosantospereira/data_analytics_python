# Simulando indicadores de saÃºde

## FunÃ§Ã£o que gera o indicador

```python
def time_series(max, n, a, b, c, date_start):
    stop = int((max - b)/a)
    x = np.linspace(0, stop, n)
    y = a * x + b + c * np.random.uniform(low = -15, high = 15, size = n) # np.random.normal(0,1,n)
    
    df_test = pd.DataFrame(y,columns=['name']).reset_index(drop=True)
    df_test['date'] = pd.date_range(start = date_start, periods = n)
    
    fig, ax = plt.subplots()
    ax.plot(x, y, label = "Data")
    return df_test
```

## Idade em anos

```python
df_age = time_series(
31,      # maximum value for the time series
350,        # number of sample size
-1,           # angular coefficient for the linear function
35,    # intercept for the linear function
0.1,         # value that multiplies the white noise
'1/1/2019'    # date to start with
)
df_age.rename(columns={'name': 'age_years'}, inplace=True)
```
